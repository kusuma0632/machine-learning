The shape of the boxplot shows how the data is distributed and it also shows any outliners
it is a uesful way to compare different sets of data as you can draw more than one boxplot per graph.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
list1=[11,4,6,8,6,9,3]
fig=plt.figure(figsize=(10,7))
plt.boxplot(list1)
plt.show()
```

```python

list2=[1,2,2,2,3,5,6]
fig=plt.figure(figsize=(10,7))
plt.boxplot(list2)
plt.show()
```

```python
list1=[11,4,6,8,6,9,3]
fig=plt.figure(figsize=(5,5))
plt.boxplot(list1,notch=True)
plt.show()
```

```python
list1=[11,4,6,8,6,9,3]
fig=plt.figure(figsize=(5,3))
plt.boxplot(list1,vert=False)
plt.show()
```

```python
list1=[11,4,6,8,6,9,3]
fig=plt.figure(figsize=(5,4))
plt.boxplot(list1,widths=0.6)
plt.show()
```

```python
list1=[11,4,6,8,6,9,3]
fig=plt.figure(figsize=(5,4))
plt.boxplot(list1,labels=["Hema"],widths=0.5)
plt.show()
```

```python
list1=[11,4,6,8,6,9,3]
fig=plt.figure(figsize=(5,4))
plt.boxplot(list1,widths=0.3,patch_artist=True)
plt.show()
```

```python
list1=[11,4,6,8,6,9,3]
fig=plt.figure(figsize=(5,4))
plt.boxplot(list1,widths=0.3,patch_artist=False,showmeans=True)
plt.show()
```

```python
list1=[11,4,6,8,6,9,3]
fig=plt.figure(figsize=(5,4))
plt.boxplot(list1,widths=0.3,patch_artist=False,showmeans=True,boxprops=dict(color="r"))
plt.show()
```

```python
list1=[11,4,6,8,6,9,3]
fig=plt.figure(figsize=(5,4))
plt.boxplot(list1,widths=0.3,patch_artist=False,showmeans=True,boxprops=dict(color="r"),capprops=dict(color="g"))
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
Student_Performance=["Excellent","Good","Average","Poor"]
Student_Values=[35,30,20,15]
plt.pie(Student_Values)
plt.show()
```

```python
Student_Performance=["Excellent","Good","Average","Poor"]
Student_Values=[35,30,20,15]
plt.pie(Student_Values,labels=Student_Performance)
plt.show()
```

```python
Student_Performance=["Excellent","Good","Average","Poor"]
Student_Values=[35,30,20,15]
plt.pie(Student_Values,labels=Student_Performance,startangle=180)
plt.show()
```

```python
Student_Performance=["Excellent","Good","Average","Poor"]
Student_Values=[35,30,20,15]
plt.pie(Student_Values,labels=Student_Performance,startangle=0,explode=[0.3,0,0,0])
plt.show()
```

```python
Student_Performance=["Excellent","Good","Average","Poor"]
Student_Values=[35,30,20,15]
plt.pie(Student_Values,labels=Student_Performance,startangle=0,explode=[0.2,0.2,0.2,0.2])
plt.show()
```

```python
Student_Performance=["Excellent","Good","Average","Poor"]
Student_Values=[35,30,20,15]
plt.pie(Student_Values,labels=Student_Performance,startangle=0,explode=[0.2,0.2,0.2,0.2],shadow=True)
plt.show()
```

```python
Student_Performance=["Excellent","Good","Average","Poor"]
Student_Values=[35,30,20,15]
plt.pie(Student_Values,labels=Student_Performance,startangle=0,explode=[0.2,0.2,0.2,0.2],colors=["pink","yellow","purple","skyblue"])
plt.legend(title="Student Performance")
plt.show()


```
