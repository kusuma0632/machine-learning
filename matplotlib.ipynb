```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,10,2)
y=x**2
print(x,y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,ls="dotted")
plt.show()
```

```python
plt.plot(x,y,ls="dashed")
plt.show()
```

```python
plt.plot(x,y,ls="dashdot")
plt.show()
```

```python
plt.plot(x,y,ls="none")
plt.show()
```

```python
plt.plot(x,y,marker=".",ls="none")
plt.show()
```

```python
plt.plot(x,y,marker=".",c="b")
plt.show()
```

```python
plt.plot(x,y,marker=".",c="skyblue",lw="10")
plt.show()
```

```python
x=np.arange(0,10,3)
y1=x**2
y2=x**3
y3=x**4
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

```python
x=[2,4,6,8,10,12]
y=[50,70,20,80,90,40]
#c=["orange","yellow","purple","black","green","maroon"]
size=[100,200,50,150,300,250]
plt.scatter(x,y,color="orange",s=size,label="3000")
x=[2,4,6,8,10,12]
y=[25,65,95,75,55,35]
#c=["orange","yellow","purple","black","green","maroon"]
size=[55,75,85,45,65,95]
plt.scatter(x,y,color="purple",s=size,label="2020")
plt.legend()
plt.title("scatter plot",color="red")
plt.show()
```

```python
marks=[90,95,20,35,70,75,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.title("Student Grade")
plt.hist(marks,grade_intervals,histtype="bar",rwidth=0.7)
plt.xticks([0,50,80,100])
plt.xlabel("Percentage")
plt.ylabel("No.of Students")
plt.show()
```

```python
marks=[90,95,20,35,70,75,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.title("Student Grade")
plt.hist(marks,grade_intervals,histtype="step",rwidth=0.7)
plt.xticks([0,50,80,100])
plt.xlabel("Percentage")
plt.ylabel("No.of Students")
plt.show()
```

```python
marks=[90,95,20,35,70,75,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.title("Student Grade")
plt.hist(marks,grade_intervals,histtype="stepfilled",rwidth=0.7,facecolor="r")
plt.xticks([0,50,80,100])
plt.xlabel("Percentage")
plt.ylabel("No.of Students")
plt.show()
```

```python
marks=[90,95,20,35,70,75,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.title("Student Grade")
plt.hist(marks,grade_intervals,histtype="bar",rwidth=0.7,facecolor="g")
plt.xticks([0,50,80,100])
plt.xlabel("Percentage")
plt.ylabel("No.of Students")
plt.show()
```
